## Create Tables

```python
%%sql
DROP TABLE IF EXISTS users;
DROP TABLE IF EXISTS movies;
DROP TABLE IF EXISTS ratings;

CREATE TABLE users (
    id INT PRIMARY KEY,
    gender VARCHAR(5),
    age INT,
    zip_code VARCHAR(255),
    occupation VARCHAR(255),
    factors VECTOR(100)
);

CREATE TABLE movies (
    id INT PRIMARY KEY,
    title VARCHAR(255),
    genres VARCHAR(255),
    poster VARCHAR(255),
    factors VECTOR(100)
);

CREATE TABLE ratings (
    user_id INT,
    movie_id INT,
    rating FLOAT,
    timestamp INT
);
```
